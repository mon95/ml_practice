# Incomplete CliffsNotes for Neural Networks

1. \(`input`\) --> \[`Neural Network`\] --> \(`output`\)
2. A neural network can be thought of as $ weird\_function(w_1, w_2, \cdots, w_n) $.
    1. __Feedforward step__: Unspeakable things are done to the input (the weights $w_i$ play a role) to compute the output.
    2. __Backpropagation__: Check how stupid the output is (error term) and punish the neural network for being dumb (use gradients, errors to change the $w_i$).
3. Yay done? Lol no. There are many intricacies in backpropagation, choosing the error function and so on.
4. NN has big math equations $ \Rightarrow $ lots of annoying problems.
5. There are many different techniques underlying backpropagation - gradient descent is just one of them and gradient descent has tiny tweaks up its sleeve (mini-batch stochastic gradient descent, adaptive learning rate and so on)
6. If you don't feel like you really understand what's going on, it's because you didn't read the link on NN. Read it.
